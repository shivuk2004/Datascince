```python
class Linear_Duplicate:
    def __init__(self):
        self.coef_=[]
        self.intercept_=[]

    def fit(self,x,y):
        g=x.to_dict()
        print(g)
        for i,j in g.items():
            xy=[]
            x2=[]
            y2=[]
            xtot=0
            ytot=0
            count=0
            for k,l in j.items():
                xtot=xtot+l
                ytot=ytot+y[count]
                xy.append(l*y[count])
                x2.append(l*l)
                y2.append(y[count]*y[count])
                count+=1

#             #coefficient=n*sum(x*y)-(sum(x)+sum(y))/n*sum(x^2)-(sum(x)*sum(y))
#             #intercept=sum(y)*sum(x^2)-sum(x)*sum(x*y)/n*sum(x^2)-sum(x)-sum(y)

            x=((count)*sum(xy)-(xtot*ytot))/((count)*sum(x2)-(xtot*xtot))
            y=(ytot*sum(x2)-xtot*sum(xy))/(count*(sum(x2))-(xtot*xtot))
            self.coef_.append(x)
            self.intercept_.append(y)

    def predict(self,*y):
        for i in y:
                b=self.coef_[0]*i+self.intercept_[0]
                print(b)
```

```python
nreg=Linear_Duplicate()
```
