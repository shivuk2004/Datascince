```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
df=pd.read_csv('music.csv')
x=df.drop(['genre'],axis=1)
y=df.genre
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=DecisionTreeClassifier()
#model.fit(x_train,y_train)
#joblib.dump(model,'musicmodel.joblib')
model=joblib.load('musicmodel.joblib')
p=model.predict(x_test)
score=accuracy_score(y_test,p)
score
```

```python
from sklearn import tree
tree.export_graphviz(model,out_file='musicgraph.dot',
                    feature_names=['age','gender'],
                    class_names=sorted(y.unique()),
                    label='all', rounded=True,filled=True)
```

```python
import numpy as np
com=['Google','Meta','Amazon','OpenAI','Microsoft']
revenue=[500,450,600,550,220]
profit=[300,280,350,325,150]
x_pos=np.arange(len(com))
plt.bar(x_pos-0.2,revenue,color='red',label="revenue",width=0.4)
plt.bar(x_pos+0.2,profit,color='green',label="profit",width=0.4)
plt.xlabel("Companies")
plt.ylabel("Revenue")
plt.title("Companies and Its performance")
plt.xticks(x_pos,com)
plt.legend()
```
